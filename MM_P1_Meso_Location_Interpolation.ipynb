```python
meso_times = np.array(meso_data['Decimal Time'])
meso_times
```

```python
lats = np.array(meso_data['Meso Center Latitude'])
```

```python
lons = np.array(meso_data['Meso Center Longitude'])
```

```python
## lat/lon interpolation ##
```

```python
interp_lats = []
interp_lons = []
for t in range(1, len(meso_times), 1):
    dt = ((meso_times[t] - meso_times[t-1]) * 3600)
    dlat = (lats[t] - lats[t - 1]) / dt
    dlon = (lons[t] - lons[t - 1]) / dt
    lat1 = lats[t - 1]
    lon1 = lons[t - 1]
    for delta_lat in range(0, dt.astype(int), 1):
        lat2 = lat1 + dlat
        interp_lats.append(lat2)
        lat1 = lat2
    for delta_lon in range(0, dt.astype(int), 1):
        lon2 = lon1 + dlon
        interp_lons.append(lon2)
        lon1 = lon2
```

```python
from matplotlib import pyplot as plt
meso_loc = pd.read_csv('/home/jupyter-josiahm/NSSL/Data/Rolling_Fork_Mesocyclone_Location.csv')

plt.scatter(interp_lons, interp_lats, color = 'cyan')
plt.scatter(x = meso_loc['Meso Center Longitude'], y = meso_loc['Meso Center Latitude'], color = 'black')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Rolling Fork Mesocyclone Center 0030 to 0200 UTC')
```

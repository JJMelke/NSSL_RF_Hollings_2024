```python
meso_times = np.array(meso_data['Decimal Time'])
```

```python
meso_times
```

```python
lats = np.array(meso_data['Meso Center Latitude'])
len(lats)
```

```python
lons = np.array(meso_data['Meso Center Longitude'])
len(lons)
```

```python
## lat/lon interpolation ##
```

```python
len(meso_times)
```

```python
interp_lats = []
interp_lons = []
for t in range(1, len(meso_times), 1):
    dt = ((meso_times[t] - meso_times[t-1]) * 3600)
    dlat = (lats[t] - lats[t - 1]) / dt
    dlon = (lons[t] - lons[t - 1]) / dt
    lat1 = lats[t - 1]
    lon1 = lons[t - 1]
    for delta_lat in range(0, dt.astype(int), 1):
        lat2 = lat1 + dlat
        interp_lats.append(lat2)
        lat1 = lat2
    for delta_lon in range(0, dt.astype(int), 1):
        lon2 = lon1 + dlon
        interp_lons.append(lon2)
        lon1 = lon2
```

```python
meso_time = np.arange((30*60+17)/3600, 1.978057, 1/3600)
#for i in range(0, len(meso_time), 1):
#    meso_time[i] = float(meso_time[i])
    
meso_interpolation = pd.DataFrame([meso_time, interp_lats, interp_lons], ['Meso Decimal Time', 'Meso Lat', 'Meso Lon'])
meso_interpolation = meso_interpolation.transpose()
meso_interpolation.to_csv('/home/jupyter-josiahm/NSSL/Data/mesodata.csv', index = False)
```

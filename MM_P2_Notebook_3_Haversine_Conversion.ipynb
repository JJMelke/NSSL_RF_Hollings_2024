```python
import pandas as pd
import numpy as np
```

```python
path = '/Users/josia/Desktop/Python/NSSL Internship/'
```

```python
data = pd.read_csv(path + 'Data/P2/5s_AVG_P2.csv')
data
```

```python
# convert to radians
```

```python
data['P2 Lat Rad'] = np.radians(np.array(data['Latitude']))
data['P2 Lon Rad'] = np.radians(data['Longitude'])
data['Meso Lat Rad'] = np.radians(data['Meso Lat'])
data['Meso Lon Rad'] = np.radians(data['Meso Lon'])
data
```

```python
# Haversine Formula
```

```python
data['a'] = ((np.sin((data['P2 Lat Rad'] - data['Meso Lat Rad']) / 2))**2 
             + np.cos(data['Meso Lat Rad'])*np.cos(data['P2 Lat Rad'])*(np.sin((data['P2 Lon Rad'] - data['Meso Lon Rad'])/2))**2)
data
```

```python
data['c'] = 2*np.arctan2(np.sqrt(data['a']), np.sqrt((1 - data['a'])))
data
```

```python
data['Distance'] = data['c']*6371 # unit for Radius of Earth is km
data
```

```python
# Bearing Formula
```

```python
y = np.sin(data['P2 Lon Rad'] - (data['Meso Lon Rad']))*np.cos(data['P2 Lat Rad'])
x = ((np.cos(data['Meso Lat Rad'])*np.sin(data['P2 Lat Rad'])) 
     - (np.sin(data['Meso Lat Rad'])*np.cos(data['P2 Lat Rad'])*np.cos(data['P2 Lon Rad'] - data['Meso Lon Rad'])))
data['Bearing'] = np.arctan2(y, x)
data
```

```python
# Convert Bearing into degrees
data['Bearing Degrees'] = (np.rad2deg(data['Bearing']))
data
```

```python
data[['Meso Lat', 'Meso Lon', 'Latitude', 'Longitude', 'Bearing Degrees']]
```

```python
data['x'] = data['Distance']*np.sin(data['Bearing'])
```

```python
data['y'] = data['Distance']*np.cos(data['Bearing'])
```

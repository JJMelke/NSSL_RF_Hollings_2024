```python
offset = 1.52
corrected_fast_temp = []
for i in range(0, len(data), 1):
    t = data['Fast Temp'][i]
    corrected_fast_temp.append((t - 1.52))
```

```python
data['Corrected Fast Temp'] = corrected_fast_temp
```

```python
theta = []
for i in range(0, len(data), 1):
    p = data['Pressure'][i]*units.hPa
    t = data['Corrected Fast Temp'][i]*units.degC
    theta_calc = metpy.calc.potential_temperature(p, t)
    theta.append(theta_calc)
```

```python
theta_e = []
for i in range(0, len(data), 1):
    p = data['Pressure'][i]*units.hPa
    t = data['Corrected Fast Temp'][i]*units.degC
    td = data['Dew Point'][i]*units.degC
    theta_e_calc = metpy.calc.equivalent_potential_temperature(p, t, td)
    theta_e.append(theta_e_calc)
```

```python
mixing_ratio = []
for i in range(0, len(data), 1):
    t = data['Fast Temp'][i]*units.degC
    rh = data[' RH Slow'][i]*units.percent
    p = data['Pressure'][i]*units.hPa
    mix_rat_calc = metpy.calc.mixing_ratio_from_relative_humidity(p, t, rh)
    mixing_ratio.append(mix_rat_calc)
```

```python
theta_v = []
molecular_weight_ratio = 0.6219569100577033
for i in range(0, len(data), 1):
    p = data['Pressure'][i]*units.hPa
    t = data['Corrected Fast Temp'][i]*units.degC
    mix = mixing_ratio[i]
    theta_v_calc = metpy.calc.virtual_potential_temperature(p, t, mix, molecular_weight_ratio)
    theta_v.append(theta_v_calc)
```

```python
wind_comp = []
for i in range (0, len(data), 1):
    spd = data['Wind Speed'][i]*units('m/s')
    dire = data['Wind Direction'][i]*units.deg
    comp = metpy.calc.wind_components(spd, dire)
    wind_comp.append(comp)
```

```python
u = []
v = []
for i in range(0, len(wind_comp), 1):
    ui, vi = wind_comp[i]
    u.append(ui)
    v.append(vi)
```

```python
storm_speed = np.repeat(23, len(data))
storm_motion = np.repeat(241, len(data))
```

```python
wind_comp_storm = []
for i in range (0, len(data), 1):
    spd = storm_speed[i]*units('m/s')
    dire = storm_motion[i]*units.deg
    comp = metpy.calc.wind_components(spd, dire)
    wind_comp_storm.append(comp)
```

```python
u_s = []
v_s = []
for i in range(0, len(wind_comp_storm), 1):
    uj, vj = wind_comp_storm[i]
    u_s.append(uj)
    v_s.append(vj)
```

```python
altimeter = [] # https://www.weather.gov/media/epz/wxcalc/altimeterSetting.pdf
a = 0.190284
b = 0.0065*1013.25**a
for i in range(0, len(data), 1):
    alt_set = ((data['Pressure'][i] - 0.3)
               *(1 + ((b/288)*((data['Altitude'][i])/((data['Pressure'][i] - 0.3)**a))))**(1/a))
    altimeter.append(alt_set)
```

```python
mslp = []
for i in range(0, len(data)):
    mslp_calc = metpy.calc.altimeter_to_sea_level_pressure(altimeter[i]*units.hPa, 
                                                           data['Altitude'][i]*units.m, 
                                                           data['Corrected Fast Temp'][i]*units.degC)
    mslp.append(mslp_calc)
```

```python
delta_d = [np.nan*units.km]
x_bar = [np.nan]
y_bar = [np.nan]

for i in range(1, len(data) - 1, 1):
    x0 = data['x'][i - 1]
    y0 = data['y'][i - 1]
    x2 = data['x'][i + 1]
    y2 = data['y'][i + 1]
    dist = (((x2 - x0)**2 + (y2 - y0)**2)**(1/2))*units.km
    x = (x2 + x0)/2
    y = (y2 + y0)/2
    delta_d.append(dist)
    x_bar.append(x)
    y_bar.append(y)
delta_d.append(np.nan*units.km)
x_bar.append(np.nan)
y_bar.append(np.nan)
```

```python
# add to df
```

```python
data['Theta'] = theta
data['Theta_e'] = theta_e
data['Theta_v'] = theta_v
#data['Vapor Pressure'] = e
data['Mixing Ratio'] = mixing_ratio
data['u_wind'] = u
data['v_wind'] = v
data['Storm Speed'] = storm_speed
data['Storm Motion'] = storm_motion
data['storm_u'] = u_s
data['storm_v'] = v_s
data['storm_rel_u'] = data['u_wind'] - data['storm_u']
data['storm_rel_v'] = data['v_wind'] - data['storm_v']
data['Altimeter'] = altimeter
data['MSLP'] = mslp
data['Delta Probe Distance'] = delta_d
data['Gradient Mean x'] = x_bar
data['Gradient Mean y'] = y_bar
```

```python
density = []
mol_weight = 0.62195691
for i in range(0, len(data), 1):
    p = data['Pressure'][i]*units.hPa
    t = data['Corrected Fast Temp'][i]*units.degC
    r = data['Mixing Ratio'][i]
    den = metpy.calc.density(p, t, r, mol_weight)
    density.append(den)
```

```python
vorticity = [np.nan*units('s**-1')]
for i in range(1, len(data) - 1, 1):
    v2 = data['storm_rel_v'][i + 1]
    v1 = data['storm_rel_v'][i - 1]
    u2 = data['storm_rel_u'][i + 1]
    u1 = data['storm_rel_u'][i - 1]
    dv = v2 - v1
    du = u2 - u2
    x1 = data['x'][i - 1]
    y1 = data['y'][i - 1]
    x2 = data['x'][i + 1]
    y2 = data['y'][i + 1]
    dx = (x2 - x1)*1000*units.m
    dy = (y2 - y1)*1000*units.m
    zeta = (dv/dx) - (du/dy)
    vorticity.append(zeta)
vorticity.append(np.nan*units('s**-1'))
```

```python
divergence = [np.nan*units('s**-1')]
for i in range(1, len(data) - 1, 1):
    v2 = data['storm_rel_v'][i + 1]
    v1 = data['storm_rel_v'][i - 1]
    u2 = data['storm_rel_u'][i + 1]
    u1 = data['storm_rel_u'][i - 1]
    dv = v2 - v1
    du = u2 - u2
    x1 = data['x'][i - 1]
    y1 = data['y'][i - 1]
    x2 = data['x'][i + 1]
    y2 = data['y'][i + 1]
    dx = (x2 - x1)*1000*units.m
    dy = (y2 - y1)*1000*units.m
    div = (du/dx) + (dv/dy)
    divergence.append(div)
divergence.append(np.nan*units('s**-1'))
```

```python
### gradient calculations ###
```

```python
grad_theta = [np.nan*units('K/km')]
for i in range(1, len(data) - 1, 1):
    theta2 = data['Theta'][i + 1]
    theta1 = data['Theta'][i - 1]
    d = data['Delta Probe Distance'][i]
    dtheta = (theta2 - theta1) / d
    grad_theta.append(dtheta)
grad_theta.append(np.nan*units('K/km'))
```

```python
grad_theta_e = [np.nan*units('K/km')]
for i in range(1, len(data) - 1, 1):
    theta_e2 = data['Theta_e'][i + 1]
    theta_e1 = data['Theta_e'][i - 1]
    d = data['Delta Probe Distance'][i]
    dtheta_e = (theta_e2 - theta_e1) / d
    grad_theta_e.append(dtheta_e)
grad_theta_e.append(np.nan*units('K/km'))
```

```python
grad_theta_v = [np.nan*units('K/km')]
for i in range(1, len(data) - 1, 1):
    theta_v2 = data['Theta_v'][i + 1]
    theta_v1 = data['Theta_v'][i - 1]
    d = data['Delta Probe Distance'][i]
    dtheta_v = (theta_v2 - theta_v1) / d
    grad_theta_v.append(dtheta_v)
grad_theta_v.append(np.nan*units('K/km'))
```

```python
grad_fast_temp = [np.nan*units('degC/km')]
for i in range(1, len(data) - 1, 1):
    temp2 = data['Corrected Fast Temp'][i + 1]*units.degC
    temp1 = data['Corrected Fast Temp'][i - 1]*units.degC
    d = data['Delta Probe Distance'][i]
    dtemp = ((temp2 - temp1) / d)
    grad_fast_temp.append(dtemp)
grad_fast_temp.append(np.nan*units('degC/km'))
```

```python
grad_slow_temp = [np.nan*units('degC/km')]
for i in range(1, len(data) - 1, 1):
    temp2 = data['Slow Temp'][i + 1]*units.degC
    temp1 = data['Slow Temp'][i - 1]*units.degC
    d = data['Delta Probe Distance'][i]
    dtemp = ((temp2 - temp1) / d)
    grad_slow_temp.append(dtemp)
grad_slow_temp.append(np.nan*units('degC/km'))
```

```python
grad_td = [np.nan*units('degC/km')]
for i in range(1, len(data) - 1, 1):
    temp2 = data['Dew Point'][i + 1]*units.degC
    temp1 = data['Dew Point'][i - 1]*units.degC
    d = data['Delta Probe Distance'][i]
    dtemp = ((temp2 - temp1) / d)
    grad_td.append(dtemp)
grad_td.append(np.nan*units('degC/km'))
```

```python
grad_rh = [np.nan*units('km**-1')]
for i in range(1, len(data) - 1, 1):
    rh2 = data[' RH Slow'][i + 1]
    rh1 = data[' RH Slow'][i - 1]
    d = data['Delta Probe Distance'][i]
    dtemp = ((rh2 - rh1) / d)
    grad_rh.append(dtemp)
grad_rh.append(np.nan*units('km**-1'))
```

```python
grad_press = [np.nan*units('hPa/km')]
for i in range(1, len(data) - 1, 1):
    press2 = data['Pressure'][i + 1]*units.hPa
    press1 = data['Pressure'][i - 1]*units.hPa
    d = data['Delta Probe Distance'][i]
    dpress = ((press2 - press1) / d)
    grad_press.append(dpress)
grad_press.append(np.nan*units('hPa/km'))
```

```python
grad_mix_rat = [np.nan*units('kg/kg/km')]
for i in range(1, len(data) - 1, 1):
    mixrat2 = data['Mixing Ratio'][i + 1]*units('kg/kg')
    mixrat1 = data['Mixing Ratio'][i - 1]*units('kg/kg')
    d = data['Delta Probe Distance'][i]
    dmix = ((mixrat2 - mixrat1) / d)
    grad_mix_rat.append(dmix)
grad_mix_rat.append(np.nan*units('kg/kg/km'))
```

```python
data['Density'] = density
data['SR Relative Vorticity'] = vorticity
data['SR Divergence'] = divergence
data['Delta Theta'] = grad_theta
data['Delta Theta_e'] = grad_theta_e
data['Delta Theta_v'] = grad_theta_v
data['Delta Fast Temp'] = grad_fast_temp
data['Delta Slow Temp'] = grad_slow_temp
data['Delta Pressure'] = grad_press
data['Delta Dew Point'] = grad_td
data['Delta RH Slow'] = grad_rh
data['Delta Mixing Ratio'] = grad_mix_rat
```

```python
# compute Baroclinic Vorticity Generation
```

```python
baro_vort = [np.nan*units('1 / s')]
for i in range(1, len(data) - 1, 1):
    gen = (((density[i + 1] - density[i - 1])/delta_d[i].to(units.m))
           *grad_press[i].to(units('Pa/m')))*(1/density[i]**2)
    gen = gen*2*units.s
    baro_vort.append(gen.to(units('1 / s')))
baro_vort.append(np.nan*units('1 / s'))
```

```python
data['Baroclinic Generated Vorticity'] = baro_vort
```

```python
# BVG Vectors #
```

```python
probe_bearing = []
x_comp = []
y_comp = []

probe_bearing.append(np.nan)


for i in range(1, len(data) - 1, 1):
    x0 = data['x'][i - 1]
    x1 = data['x'][i + 1]
    y0 = data['y'][i - 1]
    y1 = data['y'][i + 1]
    dx = x1 - x0
    dy = y1 - y0
    if (dx < 0) & (dy < 0):
        angle = np.arctan((dx/dy)) + np.pi
        probe_bearing.append(angle)
    elif (dx < 0) & (dy > 0):
        angle = np.arctan((dy/dx)) + ((3*np.pi)/2)
        probe_bearing.append(angle)
    elif (dx > 0) & (dy < 0):
        angle = np.pi - np.arctan((dx/dy))
        probe_bearing.append(angle)
    elif (dx > 0) & (dy > 0):
        angle = (np.pi/2) - np.arctan((dy/dx))
        probe_bearing.append(angle)
    elif (dx == 0) & (dy > 0):
        angle = 0
        probe_bearing.append(angle)
    elif (dx == 0) & (dy < 0):
        angle = np.pi
        probe_bearing.append(angle)
    elif (dx < 0) & (dy == 0):
        angle = (3*np.pi)/2
        probe_bearing.append(angle)
    else : # dx > 0 & dy == 0
        angle = np.pi/2
        probe_bearing.append(angle)
probe_bearing.append(np.nan)

for j in range(0, len(data) - 1, 1):
    mag = data['Baroclinic Generated Vorticity'][j]
    if (mag > 0):
        mag_bear = probe_bearing[j] - np.pi/2
        u = np.abs(mag)*np.sin(mag_bear)
        v = np.abs(mag)*np.cos(mag_bear)
        x_comp.append(u)
        y_comp.append(v)
    elif (mag < 0):
        mag_bear = probe_bearing[j] + np.pi/2
        u = np.abs(mag)*np.sin(mag_bear)
        v = np.abs(mag)*np.cos(mag_bear)
        x_comp.append(u)
        y_comp.append(v)
    else : # mag == 0
        x_comp.append(np.nan)
        y_comp.append(np.nan)

x_comp.append(np.nan)
y_comp.append(np.nan)

```

```python
data['Probe Bearing'] = probe_bearing
data['BGV x_comp'] = x_comp
data['BGV y_comp'] = y_comp
```

```python
wind_bearing = []
bgv_bearing = []

for i in range(0, len(data), 1):
    u = data['storm_rel_u'][i]
    v = data['storm_rel_v'][i]
    if (u < 0) & (v < 0):
        angle = np.arctan((u/v)) + np.pi
        wind_bearing.append(angle)
    elif (u < 0) & (v > 0):
        angle = np.arctan((v/u)) + ((3*np.pi)/2)
        wind_bearing.append(angle)
    elif (u > 0) & (v < 0):
        angle = np.pi - np.arctan((u/v))
        wind_bearing.append(angle)
    elif (u > 0) & (v > 0):
        angle = (np.pi/2) - np.arctan((v/u))
        wind_bearing.append(angle)
    elif (u == 0) & (v > 0):
        angle = 0
        wind_bearing.append(angle)
    elif (u == 0) & (v < 0):
        angle = np.pi
        wind_bearing.append(angle)
    elif (u < 0) & (v == 0):
        angle = (3*np.pi)/2
        wind_bearing.append(angle)
    else : # dx > 0 & dy == 0
        angle = np.pi/2
        wind_bearing.append(angle)

for i in range(0, len(data), 1):
    u = data['BGV x_comp'][i]
    v = data['BGV y_comp'][i]
    if (u < 0) & (v < 0):
        angle = np.arctan((u/v)) + np.pi
        bgv_bearing.append(angle)
    elif (u < 0) & (v > 0):
        angle = np.arctan((v/u)) + ((3*np.pi)/2)
        bgv_bearing.append(angle)
    elif (u > 0) & (v < 0):
        angle = np.pi - np.arctan((u/v))
        bgv_bearing.append(angle)
    elif (u > 0) & (v > 0):
        angle = (np.pi/2) - np.arctan((v/u))
        bgv_bearing.append(angle)
    elif (u == 0) & (v > 0):
        angle = 0
        bgv_bearing.append(angle)
    elif (u == 0) & (v < 0):
        angle = np.pi
        bgv_bearing.append(angle)
    elif (u < 0) & (v == 0):
        angle = (3*np.pi)/2
        bgv_bearing.append(angle)
    else : # dx > 0 & dy == 0
        angle = np.pi/2
        bgv_bearing.append(angle)

        
streamwise_vort_comp = []
crosswise_vort_comp = []

for i in range(0, len(data), 1):
    vort_mag = data['Baroclinic Generated Vorticity'][i]
    wind_angle = wind_bearing[i]
    bvg_angle = bgv_bearing[i]
    alpha = wind_angle - bvg_angle
    sw = vort_mag*np.cos(alpha)
    cw = vort_mag*np.sin(alpha)
    streamwise_vort_comp.append(sw)
    crosswise_vort_comp.append(cw)
```

```python
data['Wind Bearing from N'] = wind_bearing
data['BGV Bearing from N'] = bgv_bearing
data['Streamwise Vort Mag'] = streamwise_vort_comp
data['Crosswise Vort Mag'] = crosswise_vort_comp
```

```python
import pandas as pd
import numpy as np
```

```python
p1 = pd.read_csv('/Users/josia/Desktop/Python/NSSL Internship/Data/P1/Cleaned_Probe1.csv')
p2 = pd.read_csv('/Users/josia/Desktop/Python/NSSL Internship/Data/P2/Cleaned_Probe2.csv')
```

```python
p1_time = np.array(p1['Decimal Time'])
p1_lat = np.array(p1['Latitude'])
p1_lon = np.array(p1['Longitude'])
p1_fast_temp = np.array(p1['Fast Temp'])
p1_slow_temp = np.array(p1['Slow Temp'])
```

```python
p2_time = np.array(p2['Decimal Time'])
p2_lat = np.array(p2['Latitude'])
p2_lon = np.array(p2['Longitude'])
p2_fast_temp = np.array(p2['Fast Temp'])
p2_slow_temp = np.array(p2['Slow Temp'])
```

```python
data = pd.DataFrame([p1_time, p1_lat, p1_lon, p1_fast_temp, p1_slow_temp, 
                     p2_time, p2_lat, p2_lon, p2_fast_temp, p2_slow_temp], 
                   ['P1 Time', 'P1 Lat', 'P1 Lon', 'P1 FT', 'P1 ST', 
                    'P2 Time', 'P2 Lat', 'P2 Lon', 'P2 FT', 'P2 ST'])
data = data.transpose()
data
```

```python
data['P1 Lat Rad'] = np.radians(np.array(data['P1 Lat']))
data['P1 Lon Rad'] = np.radians(data['P1 Lon'])
data['P2 Lat Rad'] = np.radians(data['P2 Lat'])
data['P2 Lon Rad'] = np.radians(data['P2 Lon'])
data
```

```python
data['a'] = ((np.sin((data['P2 Lat Rad'] - data['P1 Lat Rad']) / 2))**2 
             + np.cos(data['P1 Lat Rad'])*np.cos(data['P2 Lat Rad'])*(np.sin((data['P2 Lon Rad'] - data['P1 Lon Rad'])/2))**2)
data
```

```python
data['c'] = 2*np.arctan2(np.sqrt(data['a']), np.sqrt((1 - data['a'])))
data
```

```python
data['Distance'] = data['c']*6371000 # unit for Radius of Earth is m
data
```

```python
p1_fast = []
p1_slow = []
p2_fast = []
p2_slow = []
time = []
for i in range(0, len(data), 1):
    if data['Distance'][i] <= 50:
        p1_fast.append(data['P1 FT'][i])
        p1_slow.append(data['P1 ST'][i])
        p2_fast.append(data['P2 FT'][i])
        p2_slow.append(data['P2 ST'][i])
        time.append(data['P1 Time'][i])
```

```python
from matplotlib import pyplot as plt
plt.scatter(time, p1_fast , color = 'red', s = 6)
plt.scatter(time, p1_slow, color = 'blue', s = 6)
plt.scatter(time, p2_fast, color = 'orange', s = 6)
plt.scatter(time, p2_slow, color = 'green', s = 6)
plt.legend(['P1 FT', 'P1 ST', 'P2 FT', 'P2 ST'])
plt.xlabel('Decimal Time')
plt.ylabel('Temperature (degC)')
plt.title('Observed Temps')
```

```python
offset = 1.52
p1_fast = np.array(p1_fast) - offset
```

```python
from matplotlib import pyplot as plt
plt.scatter(time, p1_fast , color = 'red', s = 6)
plt.scatter(time, p1_slow, color = 'blue', s = 6)
plt.scatter(time, p2_fast, color = 'orange', s = 6)
plt.scatter(time, p2_slow, color = 'green', s = 6)
plt.legend(['P1 FT', 'P1 ST', 'P2 FT', 'P2 ST'])
plt.xlabel('Decimal Time')
plt.ylabel('Temperature (degC)')
plt.title('Temps with P1 Fast Temp decreased by 1.52 degC')
```

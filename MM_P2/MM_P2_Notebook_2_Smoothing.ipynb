```python
def probe2_smooth(dt):
    # For var names, p1 is actually p2 due to p2 data file being called
    # didn't change var name to p2
    import pandas as pd
    import math
    import numpy as np
    # import metpy.calc as mpcalc
    # from metpy.plots import Hodograph, SkewT
    # from metpy.units import units
    # from metpy.units import units
    from numpy import genfromtxt
    import matplotlib.pyplot as plt
    from mpl_toolkits.axes_grid1.inset_locator import inset_axes
    # import colorcet as cc
    import cartopy.crs as ccrs
    import cartopy.feature as cfeature
    import cartopy.io.shapereader as shpreader
    import csv
    
    path = '/Users/josia/Desktop/Python/NSSL Internship/'
    
    #plot_hour = 1
    #plot_minute = 10
    #plot_second = 44
    #plot_decimal_time = float(plot_hour)+(float(plot_minute)/60.)+(float(plot_second)/3600.)
    #plot_time_string = str(plot_hour)+str(plot_minute)+str(plot_second)
    
    # probe 1 data file
    #file1 = genfromtxt('/home/jupyter-josiahm/NSSL/Data/Probe_2_IOP3_QC_all.dat',skip_header=1)

    # select variables from probe 1 data file
    #wnd_spd_der_p1 = file1[:,3]
    #wnd_dir_der_p1 = file1[:,4]
    #temp_slow_p1 = file1[:,5]
    #rh_slow_p1 = file1[:,6]
    #temp_fast_p1 = file1[:,7]
    #td_p1 = file1[:,8]
    #pressure_p1 = file1[:,10]
    #date_gps_p1 = file1[:,12]
    #time_gps_p1 = file1[:,13]
    #lat_p1 = file1[:,15]
    #lon_p1 = file1[:,16]
    #alt_p1 = file1[:,17]
    #gps_spd_p1 = file1[:,18]
    
    # convert / calculate time from gps
    #day_p1 = np.zeros(len(date_gps_p1))
    #month_p1 = np.zeros(len(date_gps_p1))
    #year_p1 = np.zeros(len(date_gps_p1))
    #hour_p1 = np.zeros(len(date_gps_p1))
    #minute_p1 = np.zeros(len(date_gps_p1))
    #second_p1 = np.zeros(len(date_gps_p1))
    #time_p1_str = []
    #for i in range(0,len(date_gps_p1)):
    #    date1 = str(date_gps_p1[i])
    #    day1 = date1[0:2]
    #    month1 = date1[2:4]
    #    year1 = date1[4:6]
    #    day_p1[i] = day1
    #    month_p1[i] = month1
    #    year_p1[i] = year1
    #    time1 = str(time_gps_p1[i])
    #    if time_gps_p1[i]<100:
    #        hour1 = 0.
    #        minute1 = 0
    #        second1 = time1
    #        hour_str = '00'
    #        minute_str = '00'
    #        if time_gps_p1[i]<10:
    #            second_str = '0'+second1
    #        else:
    #            second_str=second1
    #    elif np.logical_and(time_gps_p1[i]>=100.,time_gps_p1[i]<1000.):
    #        hour1 = 0.
    #        minute1 = time1[0]
    #        second1 = time1[1:3]
    #        hour_str = '00'
    #        minute_str = '0'+minute1
    #        second_str = second1
    #    elif np.logical_and(time_gps_p1[i]>=1000.,time_gps_p1[i]<10000.):
    #        hour1 = 0.
    #        minute1 = time1[0:2]
    #        second1 = time1[2:4]
    #        hour_str = '00'
    #        minute_str = minute1
    #        second_str = second1
    #    elif np.logical_and(time_gps_p1[i]>=10000.,time_gps_p1[i]<100000.):
    #        hour1 = time1[0]
    #        minute1 = time1[1:3]
    #        second1 = time1[3:5]
    #        hour_str = '0'+hour1
    #        minute_str = minute1
    #        second_str = second1
    #    else:
    #        hour1 = time1[0:2]
    #        minute1 = time1[2:4]
    #        second1 = time1[4:6]
    #        hour_str = hour1
    #        minute_str = minute1
    #        second_str = second1
    #    hour_p1[i] = hour1
    #    minute_p1[i] = minute1
    #    second_p1[i] = second1
    #    time_p1_str.append(hour_str+':'+minute_str+':'+second_str)
    #time_p1_str = np.array(time_p1_str)

    #decimal_time_p1 = hour_p1+(minute_p1/60.)+(second_p1/3600.)
    
    #idx, =np.where(day_p1==25)
    #decimal_time_iop_p1 = decimal_time_p1[idx]
    #wnd_spd_iop_p1 = wnd_spd_der_p1[idx]
    #wnd_dir_iop_p1 = wnd_dir_der_p1[idx]
    #pressure_iop_p1 = pressure_p1[idx]
    #temp_fast_iop_p1 = temp_fast_p1[idx]
    #td_iop_p1 = td_p1[idx]
    #alt_iop_p1 = alt_p1[idx]
    #gps_spd_iop_p1 = gps_spd_p1[idx]
    #lat_iop_p1 = lat_p1[idx]
    #lon_iop_p1 = lon_p1[idx]
    
    ### BEGIN SMOOTHING ###
        # This is where code was added by Josiah J. Melke, 29 May 2024
    # p1 is acutally p2
    
    probe1 = pd.read_csv(path + 'Data/P2/Cleaned_Probe2.csv')
    decimal_time_iop_p1 = np.array(probe1['Decimal Time'])
    lat_iop_p1 = np.array(probe1['Latitude'])
    lon_iop_p1 = np.array(probe1['Longitude'])
    temp_fast_iop_p1 = np.array(probe1['Fast Temp'])
    temp_slow_iop_p1 = np.array(probe1['Slow Temp'])
    rh_slow_iop_p1 = np.array(probe1['Slow RH'])
    td_iop_p1 = np.array(probe1['Dew Point'])
    pressure_iop_p1 = np.array(probe1['Pressure'])
    wnd_spd_iop_p1 = np.array(probe1['Wind Speed'])
    wnd_dir_iop_p1 = np.array(probe1['Wind Direction'])
    alt_iop_p1 = np.array(probe1['Altitude'])
    gps_spd_iop_p1 = np.array(probe1['GPS Speed'])
    
    # arrays of zeros for place holding
    smoothed_5s_temp_fast_iop_p1 = np.zeros(len(temp_fast_iop_p1))
    smoothed_5s_temp_fast_iop_p1[:] = np.nan
    smoothed_5s_temp_slow_iop_p1 = np.zeros(len(temp_slow_iop_p1))
    smoothed_5s_temp_slow_iop_p1[:] = np.nan
    smoothed_5s_rh_slow_iop_p1 = np.zeros(len(rh_slow_iop_p1))
    smoothed_5s_rh_slow_iop_p1[:] = np.nan
    smoothed_5s_td_iop_p1 = np.zeros(len(td_iop_p1))
    smoothed_5s_td_iop_p1[:] = np.nan
    smoothed_5s_pressure_iop_p1 = np.zeros(len(pressure_iop_p1))
    smoothed_5s_pressure_iop_p1[:] = np.nan
    smoothed_5s_wnd_spd_iop_p1 = np.zeros(len(wnd_spd_iop_p1))
    smoothed_5s_wnd_spd_iop_p1[:] = np.nan
    smoothed_5s_wnd_dir_iop_p1 = np.zeros(len(wnd_dir_iop_p1))
    smoothed_5s_wnd_dir_iop_p1[:] = np.nan
    smoothed_5s_alt_iop_p1 = np.zeros(len(alt_iop_p1))
    smoothed_5s_alt_iop_p1[:] = np.nan
    smoothed_5s_gps_spd_iop_p1 = np.zeros(len(gps_spd_iop_p1))
    smoothed_5s_gps_spd_iop_p1[:] = np.nan
    
    # number of values on each side of centered dt average
    dn = math.floor(dt/2)
    
    # j is index's in data
    # 1st index with averaged values is dn
    # loop from dn to length of data minus dn by 1's
    # ie: 5s average, dt = 5, dn = 2
    # loop from data[j] 2 to len(data) - 2
    for j in range(dn, len(temp_fast_iop_p1) - dn, 1):
        value_holder_temp_fast = []
        value_holder_temp_slow = []
        value_holder_rh_slow = []
        value_holder_td = []
        value_holder_p = []
        value_holder_wnd_spd = []
        value_holder_wnd_dir = []
        value_holder_alt = []
        value_holder_gps_spd = []
        
        # Loops through values (t) for averaging
        # Append values to [] 
        # ie: 5s average, if j = 2 (1st run of loop)
        # value_holder is array of variable indexed at 0, 1, 2, 3, 4 to give 5s average
        for t in range(j - dn, j + dn + 1, 1):
            value_holder_temp_fast.append(temp_fast_iop_p1[t])
            value_holder_temp_slow.append(temp_slow_iop_p1[t])
            value_holder_rh_slow.append(rh_slow_iop_p1[t])
            value_holder_td.append(td_iop_p1[t])
            value_holder_p.append(pressure_iop_p1[t])
            value_holder_wnd_spd.append(wnd_spd_iop_p1[t])
            value_holder_wnd_dir.append(wnd_dir_iop_p1[t])
            value_holder_alt.append(alt_iop_p1[t])
            value_holder_gps_spd.append(gps_spd_iop_p1[t])

        # Convert to np.array
        value_holder_temp_fast = np.array(value_holder_temp_fast)
        value_holder_temp_slow = np.array(value_holder_temp_slow)
        value_holder_rh_slow = np.array(value_holder_rh_slow)
        value_holder_td = np.array(value_holder_td)
        value_holder_p = np.array(value_holder_p)
        value_holder_wnd_spd = np.array(value_holder_wnd_spd)
        value_holder_wnd_dir = np.array(value_holder_wnd_dir)
        value_holder_alt = np.array(value_holder_alt)
        value_holder_gps_spd = np.array(value_holder_gps_spd)
        
        # Take mean and place in smoothed_* 
        smoothed_5s_temp_fast_iop_p1[j] = np.nanmean(value_holder_temp_fast)
        smoothed_5s_temp_slow_iop_p1[j] = np.nanmean(value_holder_temp_slow)
        smoothed_5s_rh_slow_iop_p1[j] = np.nanmean(value_holder_rh_slow)
        smoothed_5s_td_iop_p1[j] = np.nanmean(value_holder_td)
        smoothed_5s_pressure_iop_p1[j] = np.nanmean(value_holder_p)
        smoothed_5s_wnd_spd_iop_p1[j] = np.nanmean(value_holder_wnd_spd)
        smoothed_5s_wnd_dir_iop_p1[j] = np.nanmean(value_holder_wnd_dir)
        smoothed_5s_alt_iop_p1[j] = np.nanmean(value_holder_alt)
        smoothed_5s_gps_spd_iop_p1[j] = np.nanmean(value_holder_gps_spd)

    # loop done
    # Compile into DataFrame
    smoothed_p1 = pd.DataFrame(data = [decimal_time_iop_p1, lat_iop_p1, 
                                       lon_iop_p1, smoothed_5s_temp_fast_iop_p1, smoothed_5s_temp_slow_iop_p1, 
                                       smoothed_5s_rh_slow_iop_p1, smoothed_5s_td_iop_p1, smoothed_5s_pressure_iop_p1,
                                       smoothed_5s_wnd_spd_iop_p1, smoothed_5s_wnd_dir_iop_p1,
                                       smoothed_5s_alt_iop_p1, smoothed_5s_gps_spd_iop_p1, 
                                       np.array(probe1['Meso Lat']), np.array(probe1['Meso Lon'])],
                              index = ['Decimal Time', 'Latitude', 'Longitude', 'Fast Temp', 'Slow Temp', ' RH Slow',
                                       'Dew Point', 'Pressure', 'Wind Speed',
                                       'Wind Direction', 'Altitude', 'GPS Speed', 'Meso Lat', 'Meso Lon'])
    
    # transpose to have variables as columns
    smoothed_p1 = smoothed_p1.transpose()

    # remove index 0, 1, 9089, and 9090 due to smoothed values = 0
    #smoothed_p1 = smoothed_p1[dn:len(temp_fast_iop_p1) - 2]

    # save smoothed df to csv
    smoothed_p1.to_csv(path + 'Data/P2/' + str(dt) + 's_AVG_P2.csv', index = False)
    return(smoothed_p1)
```

```python
### convert time to hhmmss, add to df
```

```python
new_time = []
for i in range(0, len(meso_loc['Time (UTC)']), 1):
    time = str(meso_loc['Time (UTC)'][i])
    hh = time[0]
    mm = time[2:4]
    ss = time[5:7]
    full_time = '0' + hh + mm + ss
    new_time.append(full_time)
new_time
```

```python
meso_loc['Digit Time'] = new_time
meso_loc
```

```python
### Convert times to decimal time ###
```

```python
time = meso_loc['Digit Time']
decimal_time_meso = []
for t in range(0, len(time), 1):
    hh = time[t][0:2]
    mm = time[t][2:4]
    ss = time[t][4:6]
    hh_in_ss = int(hh)*3600
    mm_in_ss = int(mm)*60
    total_sec = hh_in_ss + mm_in_ss + int(ss)
    decimal_time_t = total_sec / 3600
    #print(decimal_time_t)
    decimal_time_meso.append(decimal_time_t)
decimal_time_meso
```

```python
### add decimal time to df ###
meso_loc['Decimal Time'] = decimal_time_meso
meso_loc
```
